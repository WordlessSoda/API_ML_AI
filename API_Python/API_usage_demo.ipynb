# 菜谱查询类API 测试
##  聚合数据菜谱大全API

```python
import requests

def DishMenu(menu)->str:
    parameters = {'menu': menu,'key': '******'}
    base = 'http://apis.juhe.cn/cook/query.php'
    response = requests.get(base, parameters)
    answer = response.json()
    return str(answer['result']['data'])

DishMenu('皮蛋瘦肉粥')
```

```python
DishMenu('八宝粥')
```

```python
DishMenu('海鲜粥')
```

##  mob菜谱查询API

```python
import requests

def DishMenu2(menu)->str:
    parameters = {'name': menu,'key': '******'}
    base = 'http://apicloud.mob.com/v1/cook/menu/search'
    response = requests.get(base, parameters)
    answer = response.json()
    return str(answer)

DishMenu2('皮蛋瘦肉粥')
```

```python
DishMenu2('八宝粥')
```

```python
DishMenu2('海鲜粥')
```

## 总结：
聚合数据的菜谱大全API相较于Mob菜谱查询，菜谱数据更加丰富，考虑到实际使用选择聚合数据的API，而不使用Mob的API。
